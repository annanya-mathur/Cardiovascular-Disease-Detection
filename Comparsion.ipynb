```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as pt
```

```python
dataframe= pd.read_csv('/content/cardio-statlog_csv.csv')
```

```python
dataframe.head(5)
```

**Relation** **between** **Diabetes** **and** **cardio**

> The graph shows linear relationship i.e. increasing case of diabetes leads to more cardiovascular problems

```python
sb.relplot("cardio","diabetes",data=dataframe,kind="line" ,color='g',estimator=sum)
pt.grid(True ,color='k')
```

**Realation between age and cardio**

```python
sb.relplot("age","cardio",data=dataframe,estimator=sum,kind="line" )
pt.grid(True ,color='k')
```

Relationship between cardio and major vessels in heart 

```python
sb.relplot("number_of_major_vessels","cardio",data=dataframe,kind="line" ,color='g')
pt.grid(True ,color='k')
```

Relationship between bp and cardio

```python
sb.relplot("resting_blood_pressure","cardio",data=dataframe,kind="line",color='g',estimator=sum)
pt.grid(True ,color='k')
```

**People** **having** **diabetes** **and** **higher** **age** **contributes** **to** **cardio** **sometimes** **but** **not** **in** **every** **case**
